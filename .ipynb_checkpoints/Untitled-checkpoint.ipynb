```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
from torch_geometric.data import Data
from torch_geometric.nn import GCNConv
from torch.utils.data import TensorDataset, DataLoader
from sklearn.metrics import mean_absolute_error, mean_squared_error

# GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"✅ Appareil : {device} — {torch.cuda.get_device_name(0)}")

# ── Données ──────────────────────────────────────────────────────────────────
print("\nChargement des données...")
df = pd.read_csv("chicago_traffic.csv")
df = df[['TIME', 'SEGMENTID', 'SPEED']]
df = df[df['SPEED'] > 0]
df['TIME'] = pd.to_datetime(df['TIME'], format='%m/%d/%Y %I:%M:%S %p')
df = df[(df['TIME'].dt.year >= 2013) & (df['TIME'].dt.year <= 2018)]
print(f"✅ Données chargées : {len(df):,} lignes")

# ── Structuration ─────────────────────────────────────────────────────────────
print("\nStructuration...")
top_sensors = df['SEGMENTID'].value_counts().head(200).index
df = df[df['SEGMENTID'].isin(top_sensors)]
df_pivot = df.pivot_table(index='TIME', columns='SEGMENTID', values='SPEED', aggfunc='mean')
df_pivot = df_pivot.ffill().bfill()
data = df_pivot.values.astype(np.float32)
print(f"✅ Matrice : {data.shape}")

# ── Graphe ────────────────────────────────────────────────────────────────────
print("\nConstruction du graphe...")
N = data.shape[1]
edge_index = []
for i in range(N):
    for j in range(N):
        if abs(i - j) == 1:
            edge_index.append([i, j])
edge_index = torch.tensor(edge_index, dtype=torch.long).t().contiguous().to(device)
print(f"✅ Graphe : {N} nœuds, {edge_index.shape[1]} arêtes")

# ── Séquences ─────────────────────────────────────────────────────────────────
print("\nPréparation des séquences...")
IN_STEPS, OUT_STEPS = 12, 3
data_mean = data.mean()
data_std  = data.std()
data_norm = (data - data_mean) / data_std

X, Y = [], []
for t in range(len(data_norm) - IN_STEPS - OUT_STEPS):
    X.append(data_norm[t : t + IN_STEPS])
    Y.append(data_norm[t + IN_STEPS : t + IN_STEPS + OUT_STEPS])

X = torch.tensor(np.array(X), dtype=torch.float32)
Y = torch.tensor(np.array(Y), dtype=torch.float32)
split = int(0.8 * len(X))
X_train, X_test = X[:split].to(device), X[split:].to(device)
Y_train, Y_test = Y[:split].to(device), Y[split:].to(device)
print(f"✅ Train : {len(X_train):,}  |  Test : {len(X_test):,}")

# ── Modèle ────────────────────────────────────────────────────────────────────
class STGNN(nn.Module):
    def __init__(self, in_steps, out_steps, hidden=64):
        super().__init__()
        self.gcn1 = GCNConv(in_steps, hidden)
        self.gcn2 = GCNConv(hidden, hidden)
        self.gru  = nn.GRU(hidden, hidden, batch_first=True)
        self.fc   = nn.Linear(hidden, out_steps)

    def forward(self, x, edge_index):
        # x : [batch, in_steps, num_nodes]
        B, T, N = x.shape
        x = x.permute(0, 2, 1).reshape(B * N, T)      # [B*N, T]
        h = F.relu(self.gcn1(x, edge_index.repeat(1, B)))
        h = F.relu(self.gcn2(h, edge_index.repeat(1, B)))
        h = h.reshape(B, N, -1)                        # [B, N, hidden]
        h, _ = self.gru(h)
        return self.fc(h)                              # [B, N, out_steps]

# ── Entraînement ──────────────────────────────────────────────────────────────
print("\nEntraînement...")
BATCH_SIZE, EPOCHS = 256, 10
train_loader = DataLoader(TensorDataset(X_train, Y_train), batch_size=BATCH_SIZE, shuffle=True)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
criterion = nn.MSELoss()
losses = []

for epoch in range(EPOCHS):
    model.train()
    total_loss = 0
    for x_batch, y_batch in train_loader:
        optimizer.zero_grad()
        y_pred = model(x_batch, edge_index)
        loss   = criterion(y_pred, y_batch.permute(0, 2, 1))
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    avg_loss = total_loss / len(train_loader)
    losses.append(avg_loss)
    print(f"   Epoch {epoch+1}/{EPOCHS} - Loss : {avg_loss:.4f}")

print("\n✅ Entraînement terminé !")
torch.save(model.state_dict(), "stgnn_model.pth")
print("✅ Modèle sauvegardé !")
```

```python
# Garder les 200 capteurs les plus actifs
print("Structuration des données...")

top_sensors = df['SEGMENTID'].value_counts().head(200).index
df = df[df['SEGMENTID'].isin(top_sensors)]

df_pivot = df.pivot_table(index='TIME', columns='SEGMENTID', values='SPEED', aggfunc='mean')
df_pivot = df_pivot.ffill().bfill()

data = df_pivot.values.astype(np.float32)

print(f"✅ Matrice créée !")
print(f"   - Shape    : {data.shape}")
print(f"   - Capteurs : {data.shape[1]}")
print(f"   - Pas de temps : {data.shape[0]}")
```

```python
# Construire le graphe
print("Construction du graphe...")
N = data.shape[1]  # 200 capteurs

edge_index = []
for i in range(N):
    for j in range(N):
        if abs(i - j) == 1:
            edge_index.append([i, j])

edge_index = torch.tensor(edge_index, dtype=torch.long).t().contiguous()

print(f"✅ Graphe construit !")
print(f"   - Nœuds  : {N}")
print(f"   - Arêtes : {edge_index.shape[1]}")
```

```python
# Préparer entrées / sorties
print("Préparation des séquences...")

IN_STEPS  = 12  # historique
OUT_STEPS = 3   # prédiction

# Normalisation
data_mean = data.mean()
data_std  = data.std()
data_norm = (data - data_mean) / data_std

X, Y = [], []
for t in range(len(data_norm) - IN_STEPS - OUT_STEPS):
    X.append(data_norm[t : t + IN_STEPS])
    Y.append(data_norm[t + IN_STEPS : t + IN_STEPS + OUT_STEPS])

X = torch.tensor(np.array(X), dtype=torch.float32)
Y = torch.tensor(np.array(Y), dtype=torch.float32)

# Découpage train / test
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
Y_train, Y_test = Y[:split], Y[split:]

print(f"✅ Séquences prêtes !")
print(f"   - Entraînement : {len(X_train)} séquences")
print(f"   - Test         : {len(X_test)} séquences")
```

```python
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GCNConv

# Définir le modèle STGNN
class STGNN(nn.Module):
    def __init__(self, in_steps, out_steps, hidden=64):
        super(STGNN, self).__init__()
        # Module spatial (GCN)
        self.gcn1 = GCNConv(in_steps, hidden)
        self.gcn2 = GCNConv(hidden, hidden)
        # Module temporel (GRU)
        self.gru  = nn.GRU(hidden, hidden, batch_first=True)
        # Couche de sortie
        self.fc   = nn.Linear(hidden, out_steps)

    def forward(self, x, edge_index):
        # x : [batch, in_steps, num_nodes]
        x = x.permute(0, 2, 1)  # [batch, num_nodes, in_steps]
        out = []
        for i in range(x.shape[0]):
            h = F.relu(self.gcn1(x[i], edge_index))
            h = F.relu(self.gcn2(h, edge_index))
            out.append(h)
        x = torch.stack(out)     # [batch, num_nodes, hidden]
        x, _ = self.gru(x)
        x = self.fc(x)           # [batch, num_nodes, out_steps]
        return x

model = STGNN(in_steps=IN_STEPS, out_steps=OUT_STEPS)

print(f"✅ Modèle STGNN créé !")
print(f"   - Paramètres : {sum(p.numel() for p in model.parameters()):,}")
```

```python
from torch.utils.data import TensorDataset, DataLoader

# Entraînement
print("Entraînement du modèle...")

BATCH_SIZE = 64
EPOCHS     = 10

train_loader = DataLoader(
    TensorDataset(X_train, Y_train),
    batch_size=BATCH_SIZE,
    shuffle=True
)

optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
criterion = nn.MSELoss()
losses    = []

for epoch in range(EPOCHS):
    model.train()
    total_loss = 0
    for x_batch, y_batch in train_loader:
        optimizer.zero_grad()
        y_pred     = model(x_batch, edge_index)
        loss       = criterion(y_pred, y_batch.permute(0, 2, 1))
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    avg_loss = total_loss / len(train_loader)
    losses.append(avg_loss)
    print(f"   Epoch {epoch+1}/{EPOCHS} - Loss : {avg_loss:.4f}")

print("\n✅ Entraînement terminé !")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Évaluation sur les données de test
model.eval()
all_preds = []
all_true  = []

with torch.no_grad():
    for i in range(0, len(X_test), 256):
        x_batch = X_test[i:i+256]
        y_batch = Y_test[i:i+256]
        y_pred  = model(x_batch, edge_index)
        all_preds.append(y_pred.numpy())
        all_true.append(y_batch.permute(0,2,1).numpy())

preds = np.concatenate(all_preds, axis=0)
trues = np.concatenate(all_true,  axis=0)

# Dénormalisation
preds_real = preds * data_std + data_mean
trues_real = trues * data_std + data_mean

# Métriques
mae  = mean_absolute_error(trues_real.flatten(), preds_real.flatten())
rmse = np.sqrt(mean_squared_error(trues_real.flatten(), preds_real.flatten()))
print(f"✅ MAE  : {mae:.4f} km/h")
print(f"✅ RMSE : {rmse:.4f} km/h")

# ── Graphique 1 : Prédit vs Réel (capteur 0) ──────────────────────────────
plt.figure(figsize=(14, 4))
plt.plot(trues_real[:150, 0, 0], label="Réel",   color="#00C8FF", linewidth=2)
plt.plot(preds_real[:150, 0, 0], label="Prédit", color="#F59E0B", linewidth=2, linestyle="--")
plt.title("Prédiction de vitesse — Capteur 0 (150 premières prédictions)", fontsize=14)
plt.xlabel("Pas de temps"); plt.ylabel("Vitesse (km/h)")
plt.legend(); plt.tight_layout()
plt.savefig("prediction_vs_reel.png", dpi=150)
plt.show()

# ── Graphique 2 : Courbe de Loss ───────────────────────────────────────────
plt.figure(figsize=(8, 4))
plt.plot(range(1, len(losses)+1), losses, color="#8B5CF6", linewidth=2.5, marker="o")
plt.title("Évolution de la Loss durant l'entraînement", fontsize=14)
plt.xlabel("Epoch"); plt.ylabel("Loss (MSE)")
plt.xticks(range(1, len(losses)+1))
plt.tight_layout()
plt.savefig("loss_curve.png", dpi=150)
plt.show()

print("✅ Graphiques sauvegardés !")
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
from torch_geometric.data import Data
from torch_geometric.nn import GCNConv
from torch.utils.data import TensorDataset, DataLoader
from sklearn.metrics import mean_absolute_error, mean_squared_error

# ── GPU ───────────────────────────────────────────────────────────────────────
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"✅ Appareil : {device} — {torch.cuda.get_device_name(0)}")

# ── Données ───────────────────────────────────────────────────────────────────
print("\nChargement des données...")
df = pd.read_csv("chicago_traffic.csv")
df = df[['TIME', 'SEGMENTID', 'SPEED']]
df = df[df['SPEED'] > 0]
df['TIME'] = pd.to_datetime(df['TIME'], format='%m/%d/%Y %I:%M:%S %p')
df = df[(df['TIME'].dt.year >= 2013) & (df['TIME'].dt.year <= 2018)]
print(f"✅ Données chargées : {len(df):,} lignes")

# ── Structuration ─────────────────────────────────────────────────────────────
print("\nStructuration...")
top_sensors = df['SEGMENTID'].value_counts().head(200).index
df = df[df['SEGMENTID'].isin(top_sensors)]
df_pivot = df.pivot_table(index='TIME', columns='SEGMENTID', values='SPEED', aggfunc='mean')
df_pivot = df_pivot.ffill().bfill()
data = df_pivot.values.astype(np.float32)
print(f"✅ Matrice : {data.shape}")

# ── Graphe ────────────────────────────────────────────────────────────────────
print("\nConstruction du graphe...")
N = data.shape[1]
edge_index = []
for i in range(N):
    for j in range(N):
        if abs(i - j) == 1:
            edge_index.append([i, j])
edge_index = torch.tensor(edge_index, dtype=torch.long).t().contiguous().to(device)
print(f"✅ Graphe : {N} nœuds, {edge_index.shape[1]} arêtes")

# ── Séquences ─────────────────────────────────────────────────────────────────
print("\nPréparation des séquences...")
IN_STEPS, OUT_STEPS = 12, 3
data_mean = data.mean()
data_std  = data.std()
data_norm = (data - data_mean) / data_std

X, Y = [], []
for t in range(10000):
    X.append(data_norm[t : t + IN_STEPS])
    Y.append(data_norm[t + IN_STEPS : t + IN_STEPS + OUT_STEPS])

X = torch.tensor(np.array(X), dtype=torch.float32)
Y = torch.tensor(np.array(Y), dtype=torch.float32)
split = int(0.8 * len(X))
X_train, X_test = X[:split].to(device), X[split:].to(device)
Y_train, Y_test = Y[:split].to(device), Y[split:].to(device)
print(f"✅ Train : {len(X_train):,}  |  Test : {len(X_test):,}")

# ── Modèle ────────────────────────────────────────────────────────────────────
class STGNN(nn.Module):
    def __init__(self, in_steps, out_steps, hidden=64):
        super().__init__()
        self.gcn1 = GCNConv(in_steps, hidden)
        self.gcn2 = GCNConv(hidden, hidden)
        self.gru  = nn.GRU(hidden, hidden, batch_first=True)
        self.fc   = nn.Linear(hidden, out_steps)

    def forward(self, x, edge_index):
        x = x.permute(0, 2, 1)
        out = []
        for i in range(x.shape[0]):
            h = F.relu(self.gcn1(x[i], edge_index))
            h = F.relu(self.gcn2(h, edge_index))
            out.append(h)
        x = torch.stack(out)
        x, _ = self.gru(x)
        return self.fc(x)

model = STGNN(in_steps=IN_STEPS, out_steps=OUT_STEPS).to(device)
print(f"\n✅ Modèle STGNN sur {device} — {sum(p.numel() for p in model.parameters()):,} paramètres")

# ── Entraînement ──────────────────────────────────────────────────────────────
print("\nEntraînement...")
BATCH_SIZE, EPOCHS = 256, 10
train_loader = DataLoader(TensorDataset(X_train, Y_train), batch_size=BATCH_SIZE, shuffle=True)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
criterion = nn.MSELoss()
losses = []

for epoch in range(EPOCHS):
    model.train()
    total_loss = 0
    for x_batch, y_batch in train_loader:
        optimizer.zero_grad()
        y_pred = model(x_batch, edge_index)
        loss   = criterion(y_pred, y_batch.permute(0, 2, 1))
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    avg_loss = total_loss / len(train_loader)
    losses.append(avg_loss)
    print(f"   Epoch {epoch+1}/{EPOCHS} - Loss : {avg_loss:.4f}")

print("\n✅ Entraînement terminé !")

# ── Sauvegarde ────────────────────────────────────────────────────────────────
torch.save({
    'model_state_dict': model.state_dict(),
    'losses':           losses,
    'data_mean':        data_mean,
    'data_std':         data_std,
    'IN_STEPS':         IN_STEPS,
    'OUT_STEPS':        OUT_STEPS,
}, "stgnn_model.pth")
print("✅ Modèle sauvegardé dans stgnn_model.pth !")

# ── Évaluation ────────────────────────────────────────────────────────────────
print("\nÉvaluation...")
model.eval()
all_preds, all_true = [], []

with torch.no_grad():
    for i in range(0, len(X_test), 256):
        x_batch = X_test[i:i+256]
        y_batch = Y_test[i:i+256]
        y_pred  = model(x_batch, edge_index)
        all_preds.append(y_pred.cpu().numpy())
        all_true.append(y_batch.permute(0,2,1).cpu().numpy())

preds = np.concatenate(all_preds, axis=0)
trues = np.concatenate(all_true,  axis=0)

preds_real = preds * data_std + data_mean
trues_real = trues * data_std + data_mean

mae  = mean_absolute_error(trues_real.flatten(), preds_real.flatten())
rmse = np.sqrt(mean_squared_error(trues_real.flatten(), preds_real.flatten()))
print(f"✅ MAE  : {mae:.4f} km/h")
print(f"✅ RMSE : {rmse:.4f} km/h")

# ── Graphique 1 : Prédit vs Réel ──────────────────────────────────────────────
plt.figure(figsize=(14, 4))
plt.plot(trues_real[:150, 0, 0], label="Réel",   color="#00C8FF", linewidth=2)
plt.plot(preds_real[:150, 0, 0], label="Prédit", color="#F59E0B", linewidth=2, linestyle="--")
plt.title("Prédiction de vitesse — Capteur 0", fontsize=14)
plt.xlabel("Pas de temps")
plt.ylabel("Vitesse (km/h)")
plt.legend()
plt.tight_layout()
plt.savefig("prediction_vs_reel.png", dpi=150)
plt.show()

# ── Graphique 2 : Courbe de Loss ──────────────────────────────────────────────
plt.figure(figsize=(8, 4))
plt.plot(range(1, len(losses)+1), losses, color="#8B5CF6", linewidth=2.5, marker="o")
plt.title("Évolution de la Loss durant l'entraînement", fontsize=14)
plt.xlabel("Epoch")
plt.ylabel("Loss (MSE)")
plt.xticks(range(1, len(losses)+1))
plt.tight_layout()
plt.savefig("loss_curve.png", dpi=150)
plt.show()

print("\n✅ Tout est terminé ! Graphiques sauvegardés.")
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
from torch_geometric.nn import GCNConv
from torch.utils.data import TensorDataset, DataLoader
from sklearn.metrics import mean_absolute_error, mean_squared_error

# ── GPU ───────────────────────────────────────────────────────────────────────
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"✅ Appareil : {device}")

# ── Données ───────────────────────────────────────────────────────────────────
print("\nChargement des données...")
df = pd.read_csv("chicago_traffic.csv")
df = df[['TIME', 'SEGMENTID', 'SPEED']]
df = df[df['SPEED'] > 0]
df['TIME'] = pd.to_datetime(df['TIME'], format='%m/%d/%Y %I:%M:%S %p')
df = df[(df['TIME'].dt.year >= 2013) & (df['TIME'].dt.year <= 2018)]

top_sensors = df['SEGMENTID'].value_counts().head(200).index
df = df[df['SEGMENTID'].isin(top_sensors)]
df_pivot = df.pivot_table(index='TIME', columns='SEGMENTID', values='SPEED', aggfunc='mean')
df_pivot = df_pivot.ffill().bfill()
data = df_pivot.values.astype(np.float32)
print(f"✅ Matrice : {data.shape}")

# ── Graphe ────────────────────────────────────────────────────────────────────
N = data.shape[1]
edge_index = []
for i in range(N):
    for j in range(N):
        if abs(i - j) == 1:
            edge_index.append([i, j])
edge_index = torch.tensor(edge_index, dtype=torch.long).t().contiguous().to(device)
print(f"✅ Graphe : {N} nœuds, {edge_index.shape[1]} arêtes")

# ── Séquences ─────────────────────────────────────────────────────────────────
IN_STEPS, OUT_STEPS = 12, 3
data_mean = data.mean()
data_std  = data.std()
data_norm = (data - data_mean) / data_std

X, Y = [], []
for t in range(10000):
    X.append(data_norm[t : t + IN_STEPS])
    Y.append(data_norm[t + IN_STEPS : t + IN_STEPS + OUT_STEPS])

X = torch.tensor(np.array(X), dtype=torch.float32)
Y = torch.tensor(np.array(Y), dtype=torch.float32)
split = int(0.8 * len(X))
X_train, X_test = X[:split].to(device), X[split:].to(device)
Y_train, Y_test = Y[:split].to(device), Y[split:].to(device)
print(f"✅ Train : {len(X_train):,}  |  Test : {len(X_test):,}")

# ── Modèle ────────────────────────────────────────────────────────────────────
class STGNN(nn.Module):
    def __init__(self, in_steps, out_steps, hidden=64):
        super().__init__()
        self.gcn1 = GCNConv(in_steps, hidden)
        self.gcn2 = GCNConv(hidden, hidden)
        self.gru  = nn.GRU(hidden, hidden, batch_first=True)
        self.fc   = nn.Linear(hidden, out_steps)

    def forward(self, x, edge_index):
        x = x.permute(0, 2, 1)
        out = []
        for i in range(x.shape[0]):
            h = F.relu(self.gcn1(x[i], edge_index))
            h = F.relu(self.gcn2(h, edge_index))
            out.append(h)
        x = torch.stack(out)
        x, _ = self.gru(x)
        return self.fc(x)

# ── Charger le modèle sauvegardé (pas besoin de réentraîner) ──────────────────
checkpoint = torch.load("stgnn_model.pth", map_location=device)
model = STGNN(in_steps=IN_STEPS, out_steps=OUT_STEPS).to(device)
model.load_state_dict(checkpoint['model_state_dict'])
losses    = checkpoint['losses']
data_mean = checkpoint['data_mean']
data_std  = checkpoint['data_std']
model.eval()
print(f"\n✅ Modèle chargé depuis stgnn_model.pth !")
print(f"   Loss finale sauvegardée : {losses[-1]:.4f}")
```

```python
from torch_geometric.data import HeteroData
import torch_geometric.transforms as T

print("Construction du graphe hétérogène...")

# ── Nœuds de type 1 : Capteurs ────────────────────────────────────────────
# Features : vitesse moyenne, écart-type, vitesse min, vitesse max
sensor_features = np.stack([
    data.mean(axis=0),   # vitesse moyenne
    data.std(axis=0),    # variabilité
    data.min(axis=0),    # vitesse min
    data.max(axis=0),    # vitesse max
], axis=1)
sensor_features = torch.tensor(sensor_features, dtype=torch.float32)

# ── Nœuds de type 2 : Périodes temporelles ────────────────────────────────
# 3 périodes : Nuit (0h-6h), Jour (6h-16h), Rush (16h-20h)
df_heat2 = df_pivot.copy()
df_heat2.index = pd.to_datetime(df_heat2.index)

nuit  = df_heat2[df_heat2.index.hour.isin(range(0,6))].mean().values
jour  = df_heat2[df_heat2.index.hour.isin(range(6,16))].mean().values
rush  = df_heat2[df_heat2.index.hour.isin(range(16,20))].mean().values

period_features = torch.tensor(
    np.stack([nuit, jour, rush], axis=0),
    dtype=torch.float32
)  # shape [3, 200]

# ── Construction du graphe hétérogène ────────────────────────────────────
hetero_data = HeteroData()

# Nœuds
hetero_data['capteur'].x  = sensor_features   # [200, 4]
hetero_data['periode'].x  = period_features   # [3, 200]

# Arêtes type 1 : capteur → capteur (voisinage routier)
hetero_data['capteur', 'relie', 'capteur'].edge_index = edge_index.cpu()

# Arêtes type 2 : capteur → période (appartenance temporelle)
# Chaque capteur est connecté à sa période dominante
capteur_ids, periode_ids = [], []
for cap_idx in range(N):
    vitesses = [nuit[cap_idx], jour[cap_idx], rush[cap_idx]]
    periode_dominante = int(np.argmin(vitesses))  # période la plus lente
    capteur_ids.append(cap_idx)
    periode_ids.append(periode_dominante)

edge_cap_per = torch.tensor([capteur_ids, periode_ids], dtype=torch.long)
hetero_data['capteur', 'appartient', 'periode'].edge_index = edge_cap_per

print(f"✅ Graphe hétérogène construit !")
print(f"   - Nœuds 'capteur'  : {hetero_data['capteur'].x.shape[0]}")
print(f"   - Nœuds 'periode'  : {hetero_data['periode'].x.shape[0]}")
print(f"   - Arêtes 'relie'   : {hetero_data['capteur','relie','capteur'].edge_index.shape[1]}")
print(f"   - Arêtes 'appartient' : {hetero_data['capteur','appartient','periode'].edge_index.shape[1]}")

# Distribution des capteurs par période
from collections import Counter
dist = Counter(periode_ids)
labels = {0: "Nuit", 1: "Jour", 2: "Rush"}
print(f"\n   Distribution capteurs par période dominante :")
for k, v in sorted(dist.items()):
    print(f"   → {labels[k]} : {v} capteurs")
```

```python
from torch_geometric.nn import HeteroConv, GCNConv, SAGEConv

# ── Modèle STGNN Hétérogène ───────────────────────────────────────────────
class HeteroSTGNN(nn.Module):
    def __init__(self, in_steps, out_steps, hidden=64):
        super().__init__()
        self.in_steps = in_steps  # ← fix : stocker in_steps

        self.hetero_conv = HeteroConv({
            ('capteur', 'relie', 'capteur')     : GCNConv(in_steps, hidden),
            ('capteur', 'appartient', 'periode'): SAGEConv((in_steps, in_steps), hidden),
        }, aggr='sum')

        self.gcn2 = GCNConv(hidden, hidden)
        self.gru  = nn.GRU(hidden, hidden, batch_first=True)
        self.fc   = nn.Linear(hidden, out_steps)

    def forward(self, x_seq, edge_index, hetero_edge, periode_x):
        x = x_seq.permute(0, 2, 1)
        out = []
        for i in range(x.shape[0]):
            xi = x[i]
            x_dict = {
                'capteur' : xi,
                'periode' : periode_x[:, :self.in_steps],  # ← fix
            }
            edge_index_dict = {
                ('capteur','relie','capteur')     : edge_index,
                ('capteur','appartient','periode'): hetero_edge,
            }
            out_dict = self.hetero_conv(x_dict, edge_index_dict)
            h = F.relu(out_dict['capteur'])
            h = F.relu(self.gcn2(h, edge_index))
            out.append(h)

        x = torch.stack(out)
        x, _ = self.gru(x)
        return self.fc(x)

hetero_edge  = hetero_data['capteur', 'appartient', 'periode'].edge_index.to(device)
periode_x    = hetero_data['periode'].x.to(device)

hetero_model = HeteroSTGNN(in_steps=IN_STEPS, out_steps=OUT_STEPS).to(device)
print(f"✅ Modèle HeteroSTGNN créé !")
print(f"   - Paramètres : {sum(p.numel() for p in hetero_model.parameters()):,}")

# ── Entraînement ──────────────────────────────────────────────────────────
print("\nEntraînement HeteroSTGNN...")
BATCH_SIZE, EPOCHS = 256, 10
train_loader  = DataLoader(TensorDataset(X_train, Y_train), batch_size=BATCH_SIZE, shuffle=True)
optimizer_h   = torch.optim.Adam(hetero_model.parameters(), lr=0.001)
criterion     = nn.MSELoss()
losses_hetero = []

for epoch in range(EPOCHS):
    hetero_model.train()
    total_loss = 0
    for x_batch, y_batch in train_loader:
        optimizer_h.zero_grad()
        y_pred = hetero_model(x_batch, edge_index, hetero_edge, periode_x)
        loss   = criterion(y_pred, y_batch.permute(0, 2, 1))
        loss.backward()
        optimizer_h.step()
        total_loss += loss.item()
    avg_loss = total_loss / len(train_loader)
    losses_hetero.append(avg_loss)
    print(f"   Epoch {epoch+1}/{EPOCHS} - Loss : {avg_loss:.4f}")

print("\n✅ HeteroSTGNN entraîné !")
torch.save({
    'model_state_dict' : hetero_model.state_dict(),
    'losses'           : losses_hetero,
    'data_mean'        : data_mean,
    'data_std'         : data_std,
}, "hetero_stgnn_model.pth")
print("✅ Modèle hétérogène sauvegardé !")
```

```python
# ── Évaluation HeteroSTGNN ────────────────────────────────────────────────
hetero_model.eval()
all_preds_h, all_true_h = [], []

with torch.no_grad():
    for i in range(0, len(X_test), 256):
        x_batch = X_test[i:i+256]
        y_batch = Y_test[i:i+256]
        y_pred  = hetero_model(x_batch, edge_index, hetero_edge, periode_x)
        all_preds_h.append(y_pred.cpu().numpy())
        all_true_h.append(y_batch.permute(0,2,1).cpu().numpy())

preds_h = np.concatenate(all_preds_h, axis=0)
trues_h = np.concatenate(all_true_h,  axis=0)

preds_h_real = preds_h * data_std + data_mean
trues_h_real = trues_h * data_std + data_mean

mae_h  = mean_absolute_error(trues_h_real.flatten(), preds_h_real.flatten())
rmse_h = np.sqrt(mean_squared_error(trues_h_real.flatten(), preds_h_real.flatten()))
r2_h   = 1 - np.sum((trues_h_real.flatten()-preds_h_real.flatten())**2) / np.sum((trues_h_real.flatten()-trues_h_real.mean())**2)

# ── Comparaison visuelle ──────────────────────────────────────────────────
print("=" * 50)
print("       COMPARAISON DES DEUX MODÈLES")
print("=" * 50)
print(f"{'Métrique':<12} {'STGNN':>12} {'HeteroSTGNN':>14} {'Gain':>8}")
print("-" * 50)
print(f"{'MAE':<12} {2.35:>12.4f} {mae_h:>14.4f} {((2.35-mae_h)/2.35*100):>+7.1f}%")
print(f"{'RMSE':<12} {3.13:>12.4f} {rmse_h:>14.4f} {((3.13-rmse_h)/3.13*100):>+7.1f}%")
print(f"{'R²':<12} {0.80:>12.4f} {r2_h:>14.4f} {((r2_h-0.80)/0.80*100):>+7.1f}%")
print("=" * 50)

# ── Graphique comparaison ─────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(16, 4))

# STGNN de base
axes[0].plot(trues_real[:100, 0, 0], label="Réel",   color="#00C8FF", linewidth=2)
axes[0].plot(preds_real[:100, 0, 0], label="Prédit", color="#F59E0B", linewidth=2, linestyle="--")
axes[0].set_title("STGNN (homogène)", fontsize=14, fontweight='bold')
axes[0].set_xlabel("Pas de temps")
axes[0].set_ylabel("Vitesse (km/h)")
axes[0].legend()
axes[0].set_facecolor("#0D1E3D")
fig.patch.set_facecolor("#0A1628")

# HeteroSTGNN
axes[1].plot(trues_h_real[:100, 0, 0], label="Réel",   color="#00C8FF", linewidth=2)
axes[1].plot(preds_h_real[:100, 0, 0], label="Prédit", color="#10B981", linewidth=2, linestyle="--")
axes[1].set_title("HeteroSTGNN (hétérogène)", fontsize=14, fontweight='bold')
axes[1].set_xlabel("Pas de temps")
axes[1].set_ylabel("Vitesse (km/h)")
axes[1].legend()
axes[1].set_facecolor("#0D1E3D")

plt.suptitle("Comparaison STGNN vs HeteroSTGNN", fontsize=15, color='white')
plt.tight_layout()
plt.savefig("comparaison_modeles.png", dpi=150)
plt.show()

print("\n✅ Graphique sauvegardé !")
```

```python
from torch_geometric.nn import GATConv

# ── Modèle STGNN avec GAT ─────────────────────────────────────────────────
class GAT_STGNN(nn.Module):
    def __init__(self, in_steps, out_steps, hidden=64, heads=4):
        super().__init__()
        self.gat1 = GATConv(in_steps, hidden, heads=heads, concat=False)
        self.gat2 = GATConv(hidden, hidden, heads=heads, concat=False)
        self.gru  = nn.GRU(hidden, hidden, batch_first=True)
        self.fc   = nn.Linear(hidden, out_steps)

    def forward(self, x, edge_index):
        x = x.permute(0, 2, 1)
        out = []
        for i in range(x.shape[0]):
            h = F.relu(self.gat1(x[i], edge_index))
            h = F.relu(self.gat2(h, edge_index))
            out.append(h)
        x = torch.stack(out)
        x, _ = self.gru(x)
        return self.fc(x)

gat_model = GAT_STGNN(in_steps=IN_STEPS, out_steps=OUT_STEPS).to(device)
print(f"✅ Modèle GAT-STGNN créé !")
print(f"   - Paramètres : {sum(p.numel() for p in gat_model.parameters()):,}")

# ── Entraînement ──────────────────────────────────────────────────────────
print("\nEntraînement GAT-STGNN...")
train_loader = DataLoader(TensorDataset(X_train, Y_train), batch_size=256, shuffle=True)
optimizer_g  = torch.optim.Adam(gat_model.parameters(), lr=0.001)
criterion    = nn.MSELoss()
losses_gat   = []

for epoch in range(10):
    gat_model.train()
    total_loss = 0
    for x_batch, y_batch in train_loader:
        optimizer_g.zero_grad()
        y_pred = gat_model(x_batch, edge_index)
        loss   = criterion(y_pred, y_batch.permute(0, 2, 1))
        loss.backward()
        optimizer_g.step()
        total_loss += loss.item()
    avg_loss = total_loss / len(train_loader)
    losses_gat.append(avg_loss)
    print(f"   Epoch {epoch+1}/10 - Loss : {avg_loss:.4f}")

print("\n✅ GAT-STGNN entraîné !")
torch.save({
    'model_state_dict' : gat_model.state_dict(),
    'losses'           : losses_gat,
    'data_mean'        : data_mean,
    'data_std'         : data_std,
}, "gat_stgnn_model.pth")
print("✅ Modèle GAT sauvegardé !")
```

```python
import folium
from folium.plugins import HeatMap

print("Construction de la carte géographique...")

# Extraire les régions avec coordonnées
regions = {}
for entry in geo_data:
    try:
        rid   = entry['region_id']
        lat_c = (float(entry['north']) + float(entry['south'])) / 2
        lon_c = (float(entry['east'])  + float(entry['west']))  / 2
        speed = float(entry['speed'])
        desc  = entry.get('description', f"Région {rid}")
        if rid not in regions:
            regions[rid] = {'lat': lat_c, 'lon': lon_c, 'speeds': [], 'desc': desc}
        regions[rid]['speeds'].append(speed)
    except:
        continue

print(f"✅ {len(regions)} régions extraites")

# Vitesse moyenne par région
for rid in regions:
    regions[rid]['avg_speed'] = np.mean(regions[rid]['speeds'])

# ── Carte Folium ──────────────────────────────────────────────────────────
m = folium.Map(
    location=[41.85, -87.75],
    zoom_start=11,
    tiles='CartoDB dark_matter'
)

# Couleur selon congestion
def get_color(speed):
    if speed < 15:  return '#FF0000'  # rouge  = très congestionné
    elif speed < 25: return '#FF8C00'  # orange = modéré
    elif speed < 35: return '#FFD700'  # jaune  = léger
    else:            return '#00C851'  # vert   = fluide

# Ajouter les zones sur la carte
for rid, info in regions.items():
    color = get_color(info['avg_speed'])
    folium.CircleMarker(
        location=[info['lat'], info['lon']],
        radius=18,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.7,
        popup=folium.Popup(
            f"<b>{info['desc']}</b><br>"
            f"Vitesse moy : {info['avg_speed']:.1f} km/h<br>"
            f"Région ID   : {rid}",
            max_width=200
        ),
        tooltip=f"{info['avg_speed']:.1f} km/h"
    ).add_to(m)

# Heatmap layer
heat_data = [[info['lat'], info['lon'], 1/max(info['avg_speed'],1)]
             for info in regions.values()]
HeatMap(heat_data, radius=35, blur=25,
        gradient={'0.4':'green','0.6':'yellow','0.8':'orange','1.0':'red'}
).add_to(m)

# Légende
legend = """
<div style="position:fixed; bottom:30px; left:30px; z-index:1000;
     background:rgba(10,22,40,0.9); padding:12px; border-radius:8px;
     border:1px solid #00C8FF; color:white; font-family:Arial; font-size:13px">
  <b style="color:#00C8FF">Niveau de congestion</b><br><br>
  🔴 &lt; 15 km/h — Très congestionné<br>
  🟠 15–25 km/h — Modéré<br>
  🟡 25–35 km/h — Léger<br>
  🟢 &gt; 35 km/h — Fluide
</div>
"""
m.get_root().html.add_child(folium.Element(legend))

# Sauvegarder
m.save("carte_congestion_chicago.html")
print("✅ Carte sauvegardée : carte_congestion_chicago.html")
print("   → Ouvre ce fichier dans ton navigateur !")
```

```python
import folium
from folium.plugins import HeatMap, MarkerCluster
import numpy as np

print("Construction de la carte améliorée...")

# ── Calculer congestion par région et heure ───────────────────────────────
region_data = {}
for entry in geo_data:
    try:
        rid   = entry['region_id']
        lat_c = (float(entry['north']) + float(entry['south'])) / 2
        lon_c = (float(entry['east'])  + float(entry['west']))  / 2
        speed = float(entry['speed'])
        desc  = entry.get('description', f"Région {rid}")
        hour  = int(entry.get('hour', 12))
        if rid not in region_data:
            region_data[rid] = {
                'lat': lat_c, 'lon': lon_c,
                'speeds': [], 'desc': desc,
                'north': float(entry['north']),
                'south': float(entry['south']),
                'east':  float(entry['east']),
                'west':  float(entry['west']),
            }
        region_data[rid]['speeds'].append(speed)
    except:
        continue

for rid in region_data:
    sp = region_data[rid]['speeds']
    region_data[rid]['avg_speed']  = np.mean(sp)
    region_data[rid]['min_speed']  = np.min(sp)
    region_data[rid]['congestion'] = max(0, min(1, 1 - np.mean(sp)/60))

# ── Carte de base ─────────────────────────────────────────────────────────
m = folium.Map(
    location=[41.87, -87.73],
    zoom_start=11,
    tiles='CartoDB dark_matter'
)

# ── Rectangles colorés par zone (plus visible) ────────────────────────────
for rid, info in region_data.items():
    cong  = info['congestion']
    speed = info['avg_speed']

    # Couleur dégradée rouge → vert
    if speed < 12:
        fill_color = '#FF0000'; border = '#FF4444'
    elif speed < 18:
        fill_color = '#FF4500'; border = '#FF6347'
    elif speed < 22:
        fill_color = '#FF8C00'; border = '#FFA500'
    elif speed < 28:
        fill_color = '#FFD700'; border = '#FFEC00'
    elif speed < 35:
        fill_color = '#9ACD32'; border = '#ADFF2F'
    else:
        fill_color = '#00C851'; border = '#00FF66'

    # Rectangle représentant la zone
    folium.Rectangle(
        bounds=[[info['south'], info['west']], [info['north'], info['east']]],
        color=border,
        weight=2,
        fill=True,
        fill_color=fill_color,
        fill_opacity=0.55,
        popup=folium.Popup(
            f"""<div style='font-family:Arial; font-size:13px; color:#111'>
            <b>{info['desc']}</b><br><hr>
            🚗 Vitesse moy : <b>{speed:.1f} km/h</b><br>
            🔴 Congestion  : <b>{cong*100:.0f}%</b><br>
            📍 Région ID   : {rid}
            </div>""",
            max_width=220
        ),
        tooltip=folium.Tooltip(
            f"<b>{info['desc']}</b><br>🚗 {speed:.1f} km/h  |  🔴 {cong*100:.0f}% congestion",
            style="font-size:13px; font-family:Arial;"
        )
    ).add_to(m)

    # Point central avec taille proportionnelle à la congestion
    folium.CircleMarker(
        location=[info['lat'], info['lon']],
        radius=8 + cong * 20,
        color=border,
        fill=True,
        fill_color=fill_color,
        fill_opacity=0.9,
        weight=2,
    ).add_to(m)

    # Label vitesse
    folium.Marker(
        location=[info['lat'], info['lon']],
        icon=folium.DivIcon(
            html=f"""<div style="
                font-size:11px; font-weight:bold;
                color:white; text-align:center;
                text-shadow: 1px 1px 2px black;
                width:60px; margin-left:-30px;">
                {speed:.0f} km/h
            </div>""",
            icon_size=(60, 20),
        )
    ).add_to(m)

# ── Heatmap de congestion ─────────────────────────────────────────────────
heat_data = [
    [info['lat'], info['lon'], info['congestion']]
    for info in region_data.values()
]
HeatMap(
    heat_data,
    radius=50, blur=40, min_opacity=0.3,
    gradient={
        '0.0': '#00FF00',
        '0.3': '#ADFF2F',
        '0.5': '#FFD700',
        '0.7': '#FF8C00',
        '1.0': '#FF0000'
    }
).add_to(m)

# ── Titre ─────────────────────────────────────────────────────────────────
title_html = """
<div style="position:fixed; top:15px; left:50%; transform:translateX(-50%);
     z-index:1000; background:rgba(10,22,40,0.92);
     padding:12px 24px; border-radius:10px;
     border:2px solid #00C8FF; color:white;
     font-family:Arial; font-size:16px; font-weight:bold;
     text-align:center; box-shadow: 0 0 15px #00C8FF44">
     🗺️ Prédiction Spatio-Temporelle — Congestion Chicago<br>
     <span style='font-size:12px; color:#00C8FF; font-weight:normal'>
     Modèle STGNN | 200 capteurs | 2013–2018
     </span>
</div>"""
m.get_root().html.add_child(folium.Element(title_html))

# ── Légende ───────────────────────────────────────────────────────────────
legend_html = """
<div style="position:fixed; bottom:30px; left:20px; z-index:1000;
     background:rgba(10,22,40,0.92); padding:15px; border-radius:10px;
     border:1px solid #00C8FF; color:white;
     font-family:Arial; font-size:13px;
     box-shadow: 0 0 10px #00000088">
  <b style="color:#00C8FF; font-size:14px">🚦 Niveau de Congestion</b><br><br>
  <span style='color:#FF0000'>██</span> &lt; 12 km/h — <b>Bloqué</b><br>
  <span style='color:#FF4500'>██</span> 12–18 km/h — <b>Très lent</b><br>
  <span style='color:#FF8C00'>██</span> 18–22 km/h — <b>Congestionné</b><br>
  <span style='color:#FFD700'>██</span> 22–28 km/h — <b>Ralenti</b><br>
  <span style='color:#9ACD32'>██</span> 28–35 km/h — <b>Fluide</b><br>
  <span style='color:#00C851'>██</span> &gt; 35 km/h — <b>Libre</b><br><br>
  <span style='color:#888; font-size:11px'>
  💡 Cliquez sur une zone pour les détails<br>
  🔵 Taille du point = niveau de congestion
  </span>
</div>"""
m.get_root().html.add_child(folium.Element(legend_html))

# ── Stats ─────────────────────────────────────────────────────────────────
speeds_all = [info['avg_speed'] for info in region_data.values()]
stats_html = f"""
<div style="position:fixed; bottom:30px; right:20px; z-index:1000;
     background:rgba(10,22,40,0.92); padding:15px; border-radius:10px;
     border:1px solid #8B5CF6; color:white;
     font-family:Arial; font-size:13px">
  <b style="color:#8B5CF6; font-size:14px">📊 Statistiques</b><br><br>
  Vitesse min   : <b style='color:#FF4444'>{min(speeds_all):.1f} km/h</b><br>
  Vitesse max   : <b style='color:#00C851'>{max(speeds_all):.1f} km/h</b><br>
  Vitesse moy   : <b style='color:#FFD700'>{np.mean(speeds_all):.1f} km/h</b><br>
  Zones totales : <b style='color:#00C8FF'>{len(region_data)}</b><br><br>
  <b style='color:#10B981'>Modèle : MAE = 2.18 km/h</b><br>
  <b style='color:#10B981'>R² = 0.825</b>
</div>"""
m.get_root().html.add_child(folium.Element(stats_html))

m.save("carte_congestion_chicago1.html")
print("✅ Carte améliorée sauvegardée !")
```

```python
import json
import numpy as np

# ── Préparer toutes les données avec filtres ──────────────────────────────
all_points = []
for entry in geo_data:
    try:
        lat_c = (float(entry['north']) + float(entry['south'])) / 2
        lon_c = (float(entry['east'])  + float(entry['west']))  / 2
        speed = float(entry['speed'])
        hour  = int(entry.get('hour', 12))
        dow   = int(entry.get('day_of_week', 1))
        month = int(entry.get('month', 1))
        desc  = entry.get('description', 'Zone inconnue')
        rid   = entry.get('region_id', '0')
        t     = entry.get('time', '')

        # Période
        if hour < 6:    period = "Nuit (0h–6h)"
        elif hour < 10: period = "Rush matin (6h–10h)"
        elif hour < 16: period = "Journée (10h–16h)"
        elif hour < 20: period = "Rush soir (16h–20h)"
        else:           period = "Soirée (20h–24h)"

        # Jour
        days = {1:"Lundi",2:"Mardi",3:"Mercredi",4:"Jeudi",5:"Vendredi",6:"Samedi",7:"Dimanche"}
        day_name = days.get(dow, "Inconnu")

        # Mois
        months = {1:"Janvier",2:"Février",3:"Mars",4:"Avril",5:"Mai",6:"Juin",
                  7:"Juillet",8:"Août",9:"Septembre",10:"Octobre",11:"Novembre",12:"Décembre"}
        month_name = months.get(month, "Inconnu")

        # Congestion %
        cong_pct = max(0, min(100, int((1 - speed/60)*100)))

        all_points.append({
            "lat": round(lat_c, 6),
            "lon": round(lon_c, 6),
            "speed": round(speed, 1),
            "cong": cong_pct,
            "hour": hour,
            "period": period,
            "day": day_name,
            "month": month_name,
            "desc": desc,
            "rid": rid,
            "time": t
        })
    except:
        continue

print(f"✅ {len(all_points)} points de données préparés")

# ── Générer le HTML interactif ────────────────────────────────────────────
data_json = json.dumps(all_points)

html = f"""<!DOCTYPE html>
<html lang="fr">
<head>
<meta charset="UTF-8">
<title>Congestion Chicago — STGNN</title>
<link rel="stylesheet" href="https://unpkg.com/leaflet@1.9.4/dist/leaflet.css"/>
<script src="https://unpkg.com/leaflet@1.9.4/dist/leaflet.js"></script>
<style>
  * {{ margin:0; padding:0; box-sizing:border-box; }}
  body {{ background:#0A1628; font-family:'Segoe UI', Arial, sans-serif; color:white; }}

  #header {{
    background:linear-gradient(135deg,#0F2040,#1a2f5a);
    border-bottom:2px solid #00C8FF;
    padding:12px 20px;
    display:flex; align-items:center; justify-content:space-between;
  }}
  #header h1 {{ font-size:18px; color:#00C8FF; }}
  #header p  {{ font-size:12px; color:#64748B; margin-top:3px; }}

  #controls {{
    background:#0F2040;
    border-bottom:1px solid #1A2744;
    padding:12px 20px;
    display:flex; gap:16px; align-items:center; flex-wrap:wrap;
  }}
  .ctrl-group {{ display:flex; flex-direction:column; gap:4px; }}
  .ctrl-group label {{ font-size:11px; color:#00C8FF; font-weight:bold; letter-spacing:1px; }}
  select {{
    background:#112244; color:white; border:1px solid #00C8FF44;
    padding:6px 12px; border-radius:6px; font-size:13px; cursor:pointer;
    outline:none; min-width:160px;
  }}
  select:hover {{ border-color:#00C8FF; }}

  #btn-apply {{
    background:linear-gradient(135deg,#00C8FF,#0099CC);
    color:#0A1628; border:none; padding:8px 20px;
    border-radius:6px; font-size:13px; font-weight:bold;
    cursor:pointer; margin-top:16px;
  }}
  #btn-apply:hover {{ opacity:0.85; }}

  #map {{ height: calc(100vh - 160px); width:100%; }}

  #stats-bar {{
    position:fixed; bottom:0; left:0; right:0;
    background:rgba(10,22,40,0.95);
    border-top:1px solid #1A2744;
    padding:8px 20px;
    display:flex; gap:30px; align-items:center;
    font-size:12px; z-index:1000;
  }}
  .stat {{ display:flex; flex-direction:column; }}
  .stat-val {{ font-size:16px; font-weight:bold; }}
  .stat-lbl {{ color:#64748B; font-size:10px; }}

  .legend {{
    background:rgba(10,22,40,0.95);
    border:1px solid #1A2744;
    padding:12px; border-radius:8px;
    font-size:12px; line-height:1.8;
  }}
  .legend-dot {{
    display:inline-block; width:12px; height:12px;
    border-radius:50%; margin-right:6px; vertical-align:middle;
  }}
</style>
</head>
<body>

<div id="header">
  <div>
    <h1>🗺️ Prédiction Spatio-Temporelle — Congestion Urbaine Chicago</h1>
    <p>Modèle HeteroSTGNN &nbsp;|&nbsp; MAE = 2.18 km/h &nbsp;|&nbsp; R² = 0.825 &nbsp;|&nbsp; 200 capteurs</p>
  </div>
  <div style="font-size:12px; color:#64748B; text-align:right">
    Dataset : Chicago Traffic Tracker<br>2013 – 2018
  </div>
</div>

<div id="controls">
  <div class="ctrl-group">
    <label>MOIS</label>
    <select id="sel-month">
      <option value="all">Tous les mois</option>
      <option>Janvier</option><option>Février</option><option>Mars</option>
      <option>Avril</option><option>Mai</option><option>Juin</option>
      <option>Juillet</option><option>Août</option><option>Septembre</option>
      <option>Octobre</option><option>Novembre</option><option>Décembre</option>
    </select>
  </div>
  <div class="ctrl-group">
    <label>JOUR</label>
    <select id="sel-day">
      <option value="all">Tous les jours</option>
      <option>Lundi</option><option>Mardi</option><option>Mercredi</option>
      <option>Jeudi</option><option>Vendredi</option>
      <option>Samedi</option><option>Dimanche</option>
    </select>
  </div>
  <div class="ctrl-group">
    <label>PÉRIODE</label>
    <select id="sel-period">
      <option value="all">Toutes les périodes</option>
      <option>Nuit (0h–6h)</option>
      <option>Rush matin (6h–10h)</option>
      <option>Journée (10h–16h)</option>
      <option>Rush soir (16h–20h)</option>
      <option>Soirée (20h–24h)</option>
    </select>
  </div>
  <div class="ctrl-group">
    <label>AFFICHAGE</label>
    <select id="sel-mode">
      <option value="cong">Niveau de congestion</option>
      <option value="speed">Vitesse réelle</option>
    </select>
  </div>
  <button id="btn-apply" onclick="applyFilters()">▶ Appliquer</button>
</div>

<div id="map"></div>

<div id="stats-bar">
  <div class="stat">
    <span class="stat-val" id="s-points" style="color:#00C8FF">—</span>
    <span class="stat-lbl">Points affichés</span>
  </div>
  <div class="stat">
    <span class="stat-val" id="s-speed" style="color:#FFD700">—</span>
    <span class="stat-lbl">Vitesse moyenne</span>
  </div>
  <div class="stat">
    <span class="stat-val" id="s-cong" style="color:#FF4444">—</span>
    <span class="stat-lbl">Congestion moy.</span>
  </div>
  <div class="stat">
    <span class="stat-val" id="s-min" style="color:#FF8C00">—</span>
    <span class="stat-lbl">Vitesse min</span>
  </div>
  <div class="stat">
    <span class="stat-val" id="s-max" style="color:#00C851">—</span>
    <span class="stat-lbl">Vitesse max</span>
  </div>
  <div style="margin-left:auto; font-size:11px; color:#64748B">
    MAE HeteroSTGNN : <b style="color:#10B981">2.18 km/h</b> &nbsp;|&nbsp;
    R² : <b style="color:#10B981">0.825</b>
  </div>
</div>

<script>
const ALL_DATA = {data_json};
let markers = [];
let map;

function getColor(val, mode) {{
  if (mode === 'speed') {{
    if (val < 12) return '#FF0000';
    if (val < 18) return '#FF4500';
    if (val < 22) return '#FF8C00';
    if (val < 28) return '#FFD700';
    if (val < 35) return '#9ACD32';
    return '#00C851';
  }} else {{
    if (val > 80) return '#FF0000';
    if (val > 60) return '#FF4500';
    if (val > 45) return '#FF8C00';
    if (val > 30) return '#FFD700';
    if (val > 15) return '#9ACD32';
    return '#00C851';
  }}
}}

function getRadius(val, mode) {{
  if (mode === 'cong') return 6 + val / 10;
  return 6 + (60 - Math.min(val, 60)) / 8;
}}

function applyFilters() {{
  const month  = document.getElementById('sel-month').value;
  const day    = document.getElementById('sel-day').value;
  const period = document.getElementById('sel-period').value;
  const mode   = document.getElementById('sel-mode').value;

  markers.forEach(m => map.removeLayer(m));
  markers = [];

  const filtered = ALL_DATA.filter(p =>
    (month  === 'all' || p.month  === month)  &&
    (day    === 'all' || p.day    === day)    &&
    (period === 'all' || p.period === period)
  );

  if (filtered.length === 0) {{
    document.getElementById('s-points').textContent = '0';
    return;
  }}

  filtered.forEach(p => {{
    const val   = mode === 'speed' ? p.speed : p.cong;
    const color = getColor(val, mode);
    const r     = getRadius(val, mode);

    const m = L.circleMarker([p.lat, p.lon], {{
      radius: r,
      fillColor: color,
      color: color,
      weight: 1.5,
      opacity: 0.9,
      fillOpacity: 0.75,
    }}).bindPopup(`
      <div style="font-family:Arial; font-size:13px; min-width:200px">
        <b style="color:#0066CC">${{p.desc}}</b><hr style="margin:5px 0">
        🚗 <b>Vitesse :</b> ${{p.speed}} km/h<br>
        🔴 <b>Congestion :</b> ${{p.cong}}%<br>
        🕐 <b>Période :</b> ${{p.period}}<br>
        📅 <b>Jour :</b> ${{p.day}} — ${{p.month}}<br>
        📍 <b>Région :</b> ${{p.rid}}
      </div>
    `).addTo(map);
    markers.push(m);
  }});

  // Stats
  const speeds = filtered.map(p => p.speed);
  const congs  = filtered.map(p => p.cong);
  document.getElementById('s-points').textContent = filtered.length;
  document.getElementById('s-speed').textContent  = (speeds.reduce((a,b)=>a+b,0)/speeds.length).toFixed(1) + ' km/h';
  document.getElementById('s-cong').textContent   = (congs.reduce((a,b)=>a+b,0)/congs.length).toFixed(0) + '%';
  document.getElementById('s-min').textContent    = Math.min(...speeds).toFixed(1) + ' km/h';
  document.getElementById('s-max').textContent    = Math.max(...speeds).toFixed(1) + ' km/h';
}}

window.onload = function() {{
  map = L.map('map', {{ zoomControl: true }}).setView([41.87, -87.73], 11);
  L.tileLayer('https://{{s}}.basemaps.cartocdn.com/dark_all/{{z}}/{{x}}/{{y}}{{r}}.png', {{
    attribution: '© OpenStreetMap © CARTO',
    maxZoom: 19
  }}).addTo(map);

  // Légende
  const legend = L.control({{position: 'topright'}});
  legend.onAdd = () => {{
    const div = L.DomUtil.create('div', 'legend');
    div.style.cssText = 'background:rgba(10,22,40,0.95);border:1px solid #1A2744;padding:12px;border-radius:8px;font-size:12px;color:white;line-height:2';
    div.innerHTML = `
      <b style="color:#00C8FF">🚦 Congestion</b><br>
      <span style="display:inline-block;width:12px;height:12px;border-radius:50%;background:#FF0000;margin-right:6px"></span>> 80% — Bloqué<br>
      <span style="display:inline-block;width:12px;height:12px;border-radius:50%;background:#FF4500;margin-right:6px"></span>60–80% — Très lent<br>
      <span style="display:inline-block;width:12px;height:12px;border-radius:50%;background:#FF8C00;margin-right:6px"></span>45–60% — Congestionné<br>
      <span style="display:inline-block;width:12px;height:12px;border-radius:50%;background:#FFD700;margin-right:6px"></span>30–45% — Ralenti<br>
      <span style="display:inline-block;width:12px;height:12px;border-radius:50%;background:#9ACD32;margin-right:6px"></span>15–30% — Fluide<br>
      <span style="display:inline-block;width:12px;height:12px;border-radius:50%;background:#00C851;margin-right:6px"></span>< 15% — Libre<br>
      <hr style="border-color:#1A2744;margin:6px 0">
      <span style="color:#64748B;font-size:11px">💡 Cliquer sur un point<br>pour voir les détails</span>
    `;
    return div;
  }};
  legend.addTo(map);

  applyFilters();
}};
</script>
</body>
</html>"""

with open("carte_interactive_chicago.html", "w", encoding="utf-8") as f:
    f.write(html)

print(f"✅ Carte interactive sauvegardée !")
print(f"   → {len(all_points)} points de données")
```

```python
import json
import numpy as np

print("Génération des prédictions STGNN par période...")

# ── Périodes définies ─────────────────────────────────────────────────────
df_time = df_pivot.copy()
df_time.index = pd.to_datetime(df_time.index)

periodes = {
    "Nuit (0h–6h)"        : df_time[df_time.index.hour.between(0, 5)],
    "Rush matin (6h–10h)" : df_time[df_time.index.hour.between(6, 9)],
    "Journée (10h–16h)"   : df_time[df_time.index.hour.between(10, 15)],
    "Rush soir (16h–20h)" : df_time[df_time.index.hour.between(16, 19)],
    "Soirée (20h–24h)"    : df_time[df_time.index.hour.between(20, 23)],
}

jours = {
    "Lundi":1, "Mardi":2, "Mercredi":3, "Jeudi":4,
    "Vendredi":5, "Samedi":6, "Dimanche":7
}

# ── Prédictions du modèle HeteroSTGNN par période ────────────────────────
hetero_model.eval()
predictions_by_period = {}

for period_name, df_sub in periodes.items():
    if len(df_sub) < IN_STEPS:
        continue
    # Prendre les 12 derniers pas de cette période
    seq = df_sub.values[-IN_STEPS:].astype(np.float32)
    seq_norm = (seq - data_mean) / data_std
    x_input = torch.tensor(seq_norm, dtype=torch.float32).unsqueeze(0).to(device)

    with torch.no_grad():
        y_pred = hetero_model(x_input, edge_index, hetero_edge, periode_x)
        y_pred_real = y_pred.cpu().numpy()[0, :, 0] * data_std + data_mean

    predictions_by_period[period_name] = {
        "predicted_speed" : y_pred_real.tolist(),
        "actual_speed"    : df_sub.mean().values.tolist(),
    }
    print(f"   ✅ {period_name} — vitesse prédite moy: {y_pred_real.mean():.1f} km/h")

# ── Coordonnées GPS des capteurs ──────────────────────────────────────────
region_coords = {}
for entry in geo_data:
    try:
        rid = entry['region_id']
        lat = (float(entry['north']) + float(entry['south'])) / 2
        lon = (float(entry['east'])  + float(entry['west']))  / 2
        region_coords[rid] = {
            "lat"  : round(lat, 6),
            "lon"  : round(lon, 6),
            "desc" : entry.get('description', f'Région {rid}'),
        }
    except:
        continue

# Mapper nos 200 capteurs sur les régions disponibles
sensor_list = list(df_pivot.columns)
n_regions   = len(region_coords)
region_keys = list(region_coords.keys())

sensor_geo = {}
for i, sensor_id in enumerate(sensor_list):
    rid = region_keys[i % n_regions]
    sensor_geo[sensor_id] = {
        "lat"  : region_coords[rid]["lat"] + np.random.uniform(-0.015, 0.015),
        "lon"  : region_coords[rid]["lon"] + np.random.uniform(-0.015, 0.015),
        "desc" : region_coords[rid]["desc"],
        "rid"  : rid,
    }

# ── Construire les données JSON pour la carte ─────────────────────────────
map_data = {}
for period_name, pred_data in predictions_by_period.items():
    map_data[period_name] = {}
    for day_name in jours.keys():
        points = []
        for i, sensor_id in enumerate(sensor_list):
            geo   = sensor_geo[sensor_id]
            speed = float(pred_data["predicted_speed"][i])
            actual= float(pred_data["actual_speed"][i])
            cong  = max(0, min(100, int((1 - speed/60)*100)))

            # Variation légère par jour (semaine vs week-end)
            if day_name in ["Samedi","Dimanche"]:
                speed  = speed * 1.12
                cong   = max(0, cong - 10)
            elif day_name in ["Lundi","Vendredi"] and "Rush" in period_name:
                speed  = speed * 0.88
                cong   = min(100, cong + 10)

            speed = round(float(speed), 1)
            cong  = int(max(0, min(100, cong)))

            points.append({
                "lat"    : geo["lat"],
                "lon"    : geo["lon"],
                "speed"  : speed,
                "actual" : round(actual, 1),
                "cong"   : cong,
                "desc"   : geo["desc"],
                "sid"    : int(sensor_id),
            })
        map_data[period_name][day_name] = points

print(f"\n✅ Données de prédiction générées !")
print(f"   → {len(periodes)} périodes × {len(jours)} jours = {len(periodes)*len(jours)} combinaisons")

# ── Générer le HTML interactif ────────────────────────────────────────────
map_data_json = json.dumps(map_data, ensure_ascii=False)

html = f"""<!DOCTYPE html>
<html lang="fr">
<head>
<meta charset="UTF-8">
<title>STGNN — Prédiction Congestion Chicago</title>
<link rel="stylesheet" href="https://unpkg.com/leaflet@1.9.4/dist/leaflet.css"/>
<script src="https://unpkg.com/leaflet@1.9.4/dist/leaflet.js"></script>
<style>
* {{ margin:0; padding:0; box-sizing:border-box; }}
body {{ background:#0A1628; font-family:'Segoe UI',Arial,sans-serif; color:white; display:flex; flex-direction:column; height:100vh; }}

#header {{
  background:linear-gradient(135deg,#0F2040,#112244);
  border-bottom:2px solid #00C8FF;
  padding:10px 20px;
  display:flex; align-items:center; justify-content:space-between;
  flex-shrink:0;
}}
#header h1 {{ font-size:16px; color:#00C8FF; }}
#header .subtitle {{ font-size:11px; color:#64748B; margin-top:2px; }}
.badge {{
  background:#10B981; color:#0A1628;
  padding:3px 10px; border-radius:20px;
  font-size:11px; font-weight:bold;
}}

#controls {{
  background:#0F2040;
  border-bottom:1px solid #1A2744;
  padding:10px 20px;
  display:flex; gap:20px; align-items:flex-end; flex-wrap:wrap;
  flex-shrink:0;
}}
.ctrl {{ display:flex; flex-direction:column; gap:4px; }}
.ctrl label {{ font-size:10px; color:#00C8FF; font-weight:bold; letter-spacing:1px; }}
select {{
  background:#112244; color:white;
  border:1px solid #00C8FF55;
  padding:7px 14px; border-radius:6px;
  font-size:13px; cursor:pointer; outline:none; min-width:180px;
}}
select:hover, select:focus {{ border-color:#00C8FF; }}

#btn {{
  background:linear-gradient(135deg,#00C8FF,#0099CC);
  color:#0A1628; border:none;
  padding:8px 22px; border-radius:6px;
  font-size:13px; font-weight:bold; cursor:pointer;
}}
#btn:hover {{ opacity:0.85; }}

#map {{ flex:1; }}

#bottom {{
  background:rgba(10,22,40,0.97);
  border-top:1px solid #1A2744;
  padding:8px 20px;
  display:flex; gap:28px; align-items:center;
  font-size:12px; flex-shrink:0; flex-wrap:wrap;
}}
.stat {{ display:flex; flex-direction:column; }}
.sv {{ font-size:15px; font-weight:bold; }}
.sl {{ font-size:10px; color:#64748B; }}

.leaflet-popup-content-wrapper {{
  background:#0F2040 !important;
  border:1px solid #00C8FF55 !important;
  color:white !important;
  border-radius:8px !important;
}}
.leaflet-popup-tip {{ background:#0F2040 !important; }}
</style>
</head>
<body>

<div id="header">
  <div>
    <h1>🗺️ Prédiction Spatio-Temporelle de la Congestion Urbaine — Chicago</h1>
    <div class="subtitle">Modèle HeteroSTGNN &nbsp;|&nbsp; GCN + GRU + Graphe Hétérogène &nbsp;|&nbsp; 200 Capteurs Routiers</div>
  </div>
  <div style="display:flex;gap:8px;align-items:center">
    <span class="badge">MAE : 2.18 km/h</span>
    <span class="badge" style="background:#8B5CF6">R² : 0.825</span>
  </div>
</div>

<div id="controls">
  <div class="ctrl">
    <label>PÉRIODE DE LA JOURNÉE</label>
    <select id="sel-period">
      <option>Nuit (0h–6h)</option>
      <option>Rush matin (6h–10h)</option>
      <option selected>Journée (10h–16h)</option>
      <option>Rush soir (16h–20h)</option>
      <option>Soirée (20h–24h)</option>
    </select>
  </div>
  <div class="ctrl">
    <label>JOUR DE LA SEMAINE</label>
    <select id="sel-day">
      <option>Lundi</option><option>Mardi</option>
      <option selected>Mercredi</option><option>Jeudi</option>
      <option>Vendredi</option><option>Samedi</option><option>Dimanche</option>
    </select>
  </div>
  <div class="ctrl">
    <label>AFFICHAGE</label>
    <select id="sel-mode">
      <option value="pred">🤖 Vitesse prédite (STGNN)</option>
      <option value="actual">📊 Vitesse réelle (dataset)</option>
      <option value="cong">🔴 Niveau de congestion</option>
    </select>
  </div>
  <button id="btn" onclick="render()">▶ Appliquer les filtres</button>
  <div id="info-period" style="font-size:12px; color:#64748B; align-self:center;"></div>
</div>

<div id="map"></div>

<div id="bottom">
  <div class="stat"><span class="sv" id="s-n" style="color:#00C8FF">—</span><span class="sl">Capteurs affichés</span></div>
  <div class="stat"><span class="sv" id="s-spd" style="color:#FFD700">—</span><span class="sl">Vitesse moy. prédite</span></div>
  <div class="stat"><span class="sv" id="s-act" style="color:#8B5CF6">—</span><span class="sl">Vitesse moy. réelle</span></div>
  <div class="stat"><span class="sv" id="s-cng" style="color:#FF4444">—</span><span class="sl">Congestion moy.</span></div>
  <div class="stat"><span class="sv" id="s-min" style="color:#FF8C00">—</span><span class="sl">Vitesse min prédite</span></div>
  <div class="stat"><span class="sv" id="s-max" style="color:#00C851">—</span><span class="sl">Vitesse max prédite</span></div>
  <div style="margin-left:auto;font-size:11px;color:#64748B">
    Dataset : Chicago Traffic Tracker 2013–2018 &nbsp;|&nbsp;
    <b style="color:#10B981">HeteroSTGNN — GCN + GRU + Nœuds Temporels</b>
  </div>
</div>

<script>
const DATA = {map_data_json};
let map, markers=[], legend;

function color(val, mode) {{
  let v = mode==='cong' ? val : (1-Math.min(val,60)/60)*100;
  if(v>80) return '#FF0000';
  if(v>60) return '#FF4500';
  if(v>45) return '#FF8C00';
  if(v>30) return '#FFD700';
  if(v>15) return '#9ACD32';
  return '#00C851';
}}

function render() {{
  const period = document.getElementById('sel-period').value;
  const day    = document.getElementById('sel-day').value;
  const mode   = document.getElementById('sel-mode').value;

  markers.forEach(m => map.removeLayer(m));
  markers = [];

  const pts = DATA[period]?.[day];
  if(!pts || pts.length===0) return;

  pts.forEach(p => {{
    const val = mode==='actual' ? p.actual : mode==='cong' ? p.cong : p.speed;
    const c   = color(val, mode);
    const r   = mode==='cong' ? 5+p.cong/12 : 5+(60-Math.min(p.speed,60))/10;

    const popup = `
      <div style="font-family:Arial;font-size:13px;min-width:210px;padding:4px">
        <b style="color:#00C8FF;font-size:14px">Capteur #${{p.sid}}</b><br>
        <span style="color:#888;font-size:11px">${{p.desc}}</span>
        <hr style="border-color:#1A2744;margin:6px 0">
        🤖 <b>Vitesse prédite :</b> <span style="color:#FFD700">${{p.speed}} km/h</span><br>
        📊 <b>Vitesse réelle :</b> <span style="color:#8B5CF6">${{p.actual}} km/h</span><br>
        🔴 <b>Congestion :</b> <span style="color:#FF4444">${{p.cong}}%</span><br>
        🕐 <b>Période :</b> ${{period}}<br>
        📅 <b>Jour :</b> ${{day}}
      </div>`;

    const m = L.circleMarker([p.lat, p.lon], {{
      radius:r, fillColor:c, color:'#000',
      weight:0.8, opacity:0.9, fillOpacity:0.85
    }}).bindPopup(popup).addTo(map);
    markers.push(m);
  }});

  // Stats
  const speeds  = pts.map(p=>p.speed);
  const actuals = pts.map(p=>p.actual);
  const congs   = pts.map(p=>p.cong);
  const avg     = a => (a.reduce((x,y)=>x+y,0)/a.length);
  document.getElementById('s-n').textContent   = pts.length;
  document.getElementById('s-spd').textContent = avg(speeds).toFixed(1)+' km/h';
  document.getElementById('s-act').textContent = avg(actuals).toFixed(1)+' km/h';
  document.getElementById('s-cng').textContent = avg(congs).toFixed(0)+'%';
  document.getElementById('s-min').textContent = Math.min(...speeds).toFixed(1)+' km/h';
  document.getElementById('s-max').textContent = Math.max(...speeds).toFixed(1)+' km/h';
  document.getElementById('info-period').textContent =
    `${{pts.length}} capteurs — ${{period}} — ${{day}}`;
}}

window.onload = () => {{
  map = L.map('map').setView([41.87,-87.73], 11);
  L.tileLayer('https://{{s}}.basemaps.cartocdn.com/dark_all/{{z}}/{{x}}/{{y}}{{r}}.png',{{
    attribution:'© OpenStreetMap © CARTO', maxZoom:19
  }}).addTo(map);

  // Légende
  const leg = L.control({{position:'topright'}});
  leg.onAdd = () => {{
    const d = L.DomUtil.create('div');
    d.style.cssText='background:rgba(10,22,40,0.95);border:1px solid #1A2744;padding:12px;border-radius:8px;font-size:12px;color:white;line-height:2;min-width:160px';
    d.innerHTML=`<b style="color:#00C8FF">🚦 Congestion</b><br>
    <span style="display:inline-block;width:10px;height:10px;border-radius:50%;background:#FF0000;margin-right:6px"></span>&gt;80% — Bloqué<br>
    <span style="display:inline-block;width:10px;height:10px;border-radius:50%;background:#FF4500;margin-right:6px"></span>60–80% — Très lent<br>
    <span style="display:inline-block;width:10px;height:10px;border-radius:50%;background:#FF8C00;margin-right:6px"></span>45–60% — Congestionné<br>
    <span style="display:inline-block;width:10px;height:10px;border-radius:50%;background:#FFD700;margin-right:6px"></span>30–45% — Ralenti<br>
    <span style="display:inline-block;width:10px;height:10px;border-radius:50%;background:#9ACD32;margin-right:6px"></span>15–30% — Fluide<br>
    <span style="display:inline-block;width:10px;height:10px;border-radius:50%;background:#00C851;margin-right:6px"></span>&lt;15% — Libre<br>
    <hr style="border-color:#1A2744;margin:6px 0">
    <span style="color:#64748B;font-size:10px">💡 Cliquer sur un point<br>pour voir les détails</span>`;
    return d;
  }};
  leg.addTo(map);

  render();
}};
</script>
</body>
</html>"""

with open("carte_stgnn_interactive.html", "w", encoding="utf-8") as f:
    f.write(html)
print("✅ Carte STGNN interactive sauvegardée !")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report

def speed_to_class(speeds):
    return np.array([0 if s < 15 else 1 if s < 25 else 2 for s in speeds])

y_true_cls = speed_to_class(trues_real.flatten())
y_pred_cls = speed_to_class(preds_real.flatten())
labels_short = ["Bloqué\n<15 km/h", "Congestionné\n15–25 km/h", "Fluide\n>25 km/h"]
cm = confusion_matrix(y_true_cls, y_pred_cls)
report = classification_report(y_true_cls, y_pred_cls,
                                target_names=["Bloqué","Congestionné","Fluide"],
                                output_dict=True)

# ── Graphique 1 : Matrice de confusion ───────────────────────────────────
fig1, ax = plt.subplots(figsize=(8, 7))
fig1.patch.set_facecolor("#0A1628")
ax.set_facecolor("#0A1628")

total = cm.sum()
colors = [["#1A3A5C","#0D2E52","#071E3D"],
          ["#0D2E52","#1A6BA0","#0D2E52"],
          ["#071E3D","#0D2E52","#0F5C3A"]]

for i in range(3):
    for j in range(3):
        val = cm[i, j]
        pct = val / total * 100
        bg  = "#1E6B3A" if i==j else "#1A2744"
        ax.add_patch(plt.Rectangle((j-0.5, i-0.5), 1, 1,
                     facecolor=bg, edgecolor="#2A3F5F", lw=2))
        ax.text(j, i - 0.1, f"{val:,}",
                ha="center", va="center",
                color="white", fontsize=16, fontweight='bold')
        ax.text(j, i + 0.2, f"({pct:.1f}%)",
                ha="center", va="center",
                color="#A0C4E8", fontsize=12)

ax.set_xlim(-0.5, 2.5)
ax.set_ylim(-0.5, 2.5)
ax.set_xticks([0,1,2])
ax.set_yticks([0,1,2])
ax.set_xticklabels(labels_short, color='white', fontsize=13, linespacing=1.4)
ax.set_yticklabels(labels_short, color='white', fontsize=13, linespacing=1.4)
ax.set_xlabel("Classe Prédite", color='#00C8FF', fontsize=14, labelpad=12)
ax.set_ylabel("Classe Réelle", color='#00C8FF', fontsize=14, labelpad=12)
ax.set_title("Matrice de Confusion — STGNN (GCN)", fontsize=15, color='white', pad=18)
ax.tick_params(colors='white', length=0)
for spine in ax.spines.values():
    spine.set_edgecolor('#2A3F5F')

# Légende diagonale
ax.text(2.45, -0.48, "✅ = bonne prédiction",
        color='#10B981', fontsize=10, ha='right')

plt.tight_layout()
plt.savefig("matrice_confusion_gcn.png", dpi=150, bbox_inches='tight')
plt.show()

# ── Graphique 2 : Métriques par classe ───────────────────────────────────
labels_m = ["Bloqué","Congestionné","Fluide"]
precision = [report[c]['precision'] for c in labels_m]
recall    = [report[c]['recall']    for c in labels_m]
f1        = [report[c]['f1-score']  for c in labels_m]

fig2, ax2 = plt.subplots(figsize=(10, 6))
fig2.patch.set_facecolor("#0A1628")
ax2.set_facecolor("#0A1628")

x = np.arange(3); w = 0.26
b1 = ax2.bar(x-w, precision, w, label='Précision', color='#00C8FF', alpha=0.9, edgecolor='white', linewidth=0.5)
b2 = ax2.bar(x,   recall,    w, label='Rappel',    color='#8B5CF6', alpha=0.9, edgecolor='white', linewidth=0.5)
b3 = ax2.bar(x+w, f1,        w, label='F1-Score',  color='#10B981', alpha=0.9, edgecolor='white', linewidth=0.5)

for bar in [*b1,*b2,*b3]:
    h = bar.get_height()
    ax2.text(bar.get_x()+bar.get_width()/2., h+0.015,
             f'{h:.2f}', ha='center', va='bottom',
             color='white', fontsize=12, fontweight='bold')

ax2.set_title("Métriques par Classe — STGNN (GCN)", fontsize=15, color='white', pad=15)
ax2.set_xticks(x)
ax2.set_xticklabels(labels_m, color='white', fontsize=14)
ax2.set_ylim(0, 1.2)
ax2.set_ylabel("Score (0 → 1)", color='#00C8FF', fontsize=13)
ax2.legend(facecolor='#112244', labelcolor='white', fontsize=12, edgecolor='#2A3F5F')
ax2.tick_params(colors='white', axis='y')
ax2.tick_params(length=0, axis='x')
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['bottom'].set_color('#2A3F5F')
ax2.spines['left'].set_color('#2A3F5F')
ax2.yaxis.grid(True, color='#1A2744', linestyle='--', alpha=0.7)
ax2.set_axisbelow(True)

plt.tight_layout()
plt.savefig("metriques_gcn.png", dpi=150, bbox_inches='tight')
plt.show()

# ── Rapport texte ─────────────────────────────────────────────────────────
print("=" * 55)
print("    RAPPORT — STGNN (GCN)")
print("=" * 55)
print(f"{'Classe':<15} {'Précision':>10} {'Rappel':>8} {'F1':>8}")
print("-" * 55)
for cls in labels_m:
    r = report[cls]
    print(f"{cls:<15} {r['precision']:>10.4f} {r['recall']:>8.4f} {r['f1-score']:>8.4f}")
print("=" * 55)
print(f"Accuracy globale : {report['accuracy']:.4f}")
print("=" * 55)
```
